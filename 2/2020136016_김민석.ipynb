```python
import matplotlib.pyplot as plt
import numpy as np
```

# 실습 1

```python
def cal_distance(x, y):
    return np.sqrt(np.power(x[0]-y[0],2)+np.power(x[1]-y[1],2))


x = np.array([-3, 2])
y = np.array([3, 2])
distance = cal_distance(x, y)

# plot 코드를 작성하시오
#x_plt = np.array(-3.0, 2.0, 2.0)
#y_plt = np.array(3.0, 2.0, 0.0)
# x_plt = [-3,0,3]
# y_plt = [2,0,2]

# plt.plot([0,-3],[0,  2])
# plt.plot([0, 3],[0,  2])
# plt.plot([-3, 3], [2, 2])
#plt.plot(x_plt, y_plt)

plt.arrow(0, 0, -3, 2, head_width = 0.2, head_length = 0.2, color = 'red', label = 'vector x')
plt.arrow(0, 0, 3, 2, head_width = 0.2, head_length = 0.2, color = 'blue', label = 'vector y')
plt.hlines(2,-3,3,linestyle = '--', color = 'green', label = '||x+y||')


plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xticks(ticks=np.arange(-4, 4, step=1))
plt.yticks(ticks=np.arange(-4, 4, step=1))
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.hlines(0,-4, 4, linestyle = '--', color = 'gray')
plt.vlines(0,4,-4,linestyle = '--', color= 'gray')
plt.legend()
plt.grid()
plt.show()

print(f'두 벡터 사이의 거리는 {distance}')
```

# 실습 2

```python
def cal_cos_theta(x, y):
    return  np.dot(x,y) / (np.sqrt(np.power(x[0], 2) + np.power(x[1], 2)) * np.sqrt(np.power(y[0], 2) + np.power(y[1],2)))


x = np.array([-3, 2])
y = np.array([3, 2])
x_unit = np.array([x[0] / np.sqrt(np.power(x[0],2) +np.power(x[1],2) ),x[1]/np.sqrt(np.power(x[0],2) +np.power(x[1], 2))])
y_unit = np.array([y[0] / np.sqrt(np.power(y[0],2) +np.power(y[1],2) ),y[1]/np.sqrt(np.power(y[0],2) +np.power(y[1], 2))])

cos_theta_1 = cal_cos_theta(x, y)
cos_theta_2 = cal_cos_theta(x_unit, y_unit)



# # plot 코드를 작성하시오
plt.arrow(0, 0, x[0], x[1], head_width = 0.2, head_length = 0.2, color = 'red', label = 'vector x')
plt.arrow(0, 0, y[0], y[1], head_width = 0.2, head_length = 0.2, color = 'blue', label = 'vector y')
plt.arrow(0, 0, x_unit[0], x_unit[1], head_width = 0.2, head_length = 0.2, color = 'pink', label = 'vector x_')
plt.arrow(0, 0, y_unit[0], y_unit[1], head_width = 0.2, head_length = 0.2, color = 'cyan', label = 'vector y_')


plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xticks(ticks=np.arange(-4, 4, step=1))
plt.yticks(ticks=np.arange(-4, 4, step=1))
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.hlines(0,-4, 4, linestyle = '--', color = 'gray')
plt.vlines(0,4,-4,linestyle = '--', color= 'gray')
plt.legend()
plt.grid()
plt.show()

print(f'x, y 사이의 cos theta 값은 {cos_theta_1}')
print(f'x_, y_ 사이의 cos theta 값은 {cos_theta_2}')
# a = cal_cos_theta(np.array([1,0]),np.array([1, np.sqrt(3)]))
# print(a)
```

# 실습 3

```python
def projection(origin_v, target_v) :
  theta = cal_cos_theta(origin_v, target_v)
  o = np.array([0,0])
  proj_size = cal_distance(o, origin_v) * theta
  unit_target = np.array([target_v[0] / cal_distance(target_v, o), target_v[1] / cal_distance(target_v, o)])
  return np.array([unit_target[0] * proj_size, unit_target[1] * proj_size])

origin_v = np.array([3, 4])
target_v = np.array([4, 0])
proj_v = projection(origin_v, target_v)

# o = np.array([0,0])
# unit_target = np.array([target_v[0]/cal_distance(target_v,o),target_v[1]/cal_distance(target_v,o)])



# plot 코드를 작성하시오

plt.arrow(0, 0, origin_v[0], origin_v[1], head_width = 0.2, head_length = 0.2, color = 'red', label = 'original vector')
plt.arrow(0, 0, target_v[0], target_v[1], head_width = 0.2, head_length = 0.2, color = 'blue', label = 'target vector')
plt.arrow(0, 0, proj_v[0], proj_v[1], head_width = 0.2, head_length = 0.2, color = 'green', label = 'projected vectro')
# plt.arrow(0, 0, unit_target[0], unit_target[1], head_width = 0.2, head_length = 0.2, color = 'pink', label = 'vectro')

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xticks(ticks=np.arange(-1, 7, step=1))
plt.yticks(ticks=np.arange(-1, 7, step=1))
plt.xlim(-1, 7)
plt.ylim(-1, 7)
plt.hlines(0,-7, 7, linestyle = '--', color = 'gray')
plt.vlines(0,7,-7,linestyle = '--', color= 'gray')
plt.legend()
plt.grid()
plt.show()


print(f'origin_vector:', origin_v)
print(f'target_vector:', target_v)
print(f'projected vector: ', proj_v)
```
