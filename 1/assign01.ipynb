```python
import IPython.display as ipd
import librosa
import librosa.display
import numpy as np
from random import *

filename = "/content/drive/MyDrive/Colab Notebooks/sound.wav" # google colab 용
# filename = "sound.wav"
data,sample_rate = librosa.load(filename)
ipd.Audio(filename)
```

```python
print('data :', data[:5])
print('data size : ', len(data))
```

```python
cf = [0, 0, 0, 0 ,0]

def func(cf) : #f(x)를 코드로 나타낸 것.
    time = np.linspace(0, 0.5, len(data)) #0부터 0.5를 datasize만큼 쪼개준다.
    sound = np.sum(cf[0]*np.sin(np.array( [i*time for i in cf[1:]] )), axis=0)
    return sound

def mse(func):
    error=np.abs(data-func()) #오차
    return np.mean(error)*1500 #왜 1500을 해야할까?
                               #mean은 산술평균을 계산해줌

def test() :
    return func(cf)

print('func error : ', mse(test))
```
