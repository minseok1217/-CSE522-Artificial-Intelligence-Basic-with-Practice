```python
class Counter(dict):
    def __init__(self, c=None):
        if c is not None:
            for k, v in c.items():
                self.increment(k, v)

    def increment(self, k, v=1):
        o = self.get(k)
        if o == None:
            self[k] = v
        else:
            self[k] = o + v

    def get(self, k):
        r = super().get(k)
        # print(r)
        if r == None:
            r = 0
        return r

    def scale(self, c):
        for k, v in self.items():
            self[k] = v * c * 1.0
            
    def sum(self):
        r = 0
        for k, v in self.items():
            r += v
        return r
    
    def normalize(self):
        self.scale(1.0 / self.sum())
```

### 문제 1

```python
def get_occurrences(f): 
    """
    Returns:
        _type_: dictionary, [key(scalar): value(list)]
    """
    r = {}
    x = data[f]
    for i, v in enumerate(x.values):
        o = r.get(v)
        if o == None:
            o = []
            r[v] = o
        o.append(i)
    return r

def get_value_counts(f):
    """
    Returns:
        _type_: dictionary, [key(scalar): value(scalar)]
    """
    r = Counter() # 필요시 사용 (사용하지 않아도 됨)
    for k, v in get_occurrences(f).items():
        r.increment(k,len(v))
    return r

column_name1 = 'age'
column_name2 = 'studytime'

print(f"Occurrences for each value of '{column_name1}': ", get_occurrences(column_name1))
print(f"Occurrences for each value of '{column_name2}': ", get_occurrences(column_name2))
# print(print(type(get_occurrences(column_name1))))

print(f"Num of occurrences for each value of '{column_name1}': ", get_value_counts(column_name1))
print(f"Num of occurrences for each value of '{column_name2}': ", get_value_counts(column_name2))
```

### 문제 2

```python
def get_pmf(f):
    """
    Returns:
        _type_: dictionary [key(scalar): value(scalar)]
    """
    r = get_value_counts(f)
    # print(r)
    r.normalize() # 확률을 구해주는 식
    return r

def get_mean(f):
    """
    Returns:
        _type_: scalar
    """
    r = 0.0
    count_r = 0
    for k, v in get_value_counts(f).items():
        # print(k)
        r += k * v
        count_r += v
    return r / count_r

def get_variance(f):
    """
    Returns:
        _type_: scalar
    """
    r = 0.0
    count = 0
    for k, v in get_value_counts(f).items():
        # print(k)
        count += v
        r += math.pow((k - get_mean(f)), 2) * v
    return r / count

for f in numerical:
    print(f'{f} = {get_mean(f)} {get_variance(f)} {get_pmf(f)}')
```

```python
column_name1 = 'studytime'
column_name2 = 'G3' # final grade

def get_covariance(f1, f2):
    """
    Returns:
        _type_: scalar
    """
    # get_value_counts(f).items()
    s = data[column_name1]
    g = data[column_name2]
    s_mean = get_mean(f1)
    g_mean = get_mean(f2)
    r = 0.0
    
    for i, j in zip(s,g):
        r += i * j 
    return r / len(data) - s_mean * g_mean #까먹을 수도 있으니 기억하고 있기!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

print(f'cov = {get_covariance(column_name1, column_name2)}')
print(f'cov = {data[column_name1].cov(data[column_name2])}') # 정답 값

def get_correlation(f1, f2):
    """                                                                                                                                                                                                                                       
    Returns:
        _type_: scalar
    """
    r = get_covariance(f1, f2)/math.sqrt(get_variance(f1)*get_variance(f2))
    return r

print(f'cor = {get_correlation(column_name1, column_name2)}')
# print("---------------")
print(f'cor = {data[column_name1].corr(data[column_name2])}') # 정답 값
# print("---------------------------------")
# print(f'cor = {data[column_name1].cov(data[column_name2])/math.sqrt(get_variance(column_name1)*get_variance(column_name2))}')
```

```python
# 참고: 시뮬레이션을 통해 측정한 확률

import random

def bernoulli(p):
    sample = 1 if random.random() < p else 0
    return sample

def binomial(n, p):
    successes = 0
    for _ in range(n):
        successes += bernoulli(p)
    return successes

n_value_a, n_value_b = 200, 100
p_value_a, p_value_b = 0.1, 0.4
data_a = [binomial(n_value_a, p_value_a) for _ in range(10000)]
data_b = [binomial(n_value_b, p_value_b) for _ in range(10000)]
data = [a+b for a, b in zip(data_a, data_b)]

# print(f'data_a: {data_a[:10]}')
# print(f'data_b: {data_b[:10]}')
# print(f'data: {data[:10]}')

mask = [1 if d >= 55 else 0 for d in data]

print(f'P(A+B>=55)={sum(mask)/len(data)}')
```
